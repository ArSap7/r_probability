Давайте поиграем в кризис! Представим себе фондовый рынок. Пусть динамика акций на нём описывается случайными блужданиями. То есть цена в текущий момент времени определяется как 

$$
y_t = y_{t-1} + \varepsilon_t.
$$

При этом значение $y_0$, из которого стартует симуляция, равно $20$. Будем считать, что когда акция выпускается на рынок, она стоит $20\$$. Также будем считать, что $\varepsilon_t \sim N(0, 3^2)$. 

Получается, что цена акции сегодня - это вчерашняя цена + какой-то случайный шум. Для полного понимания того, как устроен процесс, давайте ссимулируем как ведёт себя одна ценная бумага.

```r
n_obs = 100
y = rep(20, n_obs)

for(i in 2:n_obs){
    # максимум берём, чтобы не было отрицательных цен 
    y[i] = max(0, y[i-1] + rnorm(1, mean=0, sd=3))
}

qplot(1:n_obs, y, geom='line')
```

Хватит глупостей. Поговорим о чуть более серьёзных вещах. Пусть у нас в распоряжении есть две ценные бумаги, поведение которых описывается случайными блужданиями. 

\begin{equation*}
\begin{aligned}
& y_t = y_{t-1} + \varepsilon_t \\
& x_t = x_{t-1} + u_t
\end{aligned}
\end{equation*}

Пусть эти два случайных блуждания независимы. То есть $\varepsilon_t \sim N(0, 3^2)$, $u_t \sim N(0, 3^2)$ и $Corr(u_t, \varepsilon_t) = 0$. 

__[4]__ Предположим, что нас интересует $Corr(x_t, y_t)$. Ясно, что теоретическая корреляция равна нулю. Обычно мы оцениваем теоретическую корреляцию с помощью выборочного коэффициента корреляции. В R его позволяет посчитать функция `cor(x, y)`. Правда ли, что эта оценка является несостоятельной? К чему она должна сходиться? К чему она сходится? Проверьте это с помощью симуляций.
