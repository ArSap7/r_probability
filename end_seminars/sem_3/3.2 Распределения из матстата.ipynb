```r
library("ggplot2")  # Пакет для красивых графиков 
library("grid")     # Пакет для субплотов

# Отрегулируем размер картинок, которые будут выдаваться в нашей тетрадке
library('repr')
options(repr.plot.width=6, repr.plot.height=3)
```

## 2. Распределение Хи-квадрат

```r
options(repr.plot.width=6, repr.plot.height=3)

ggplot(NULL, aes(c(0,15)))+
    stat_function(fun = dchisq, args = c(df=2), aes(color = 'Chi(2)'), size = 1) + 
    stat_function(fun = dchisq, args = c(df=3), aes(color = 'Chi(3)'), size = 1) +
    stat_function(fun = dchisq, args = c(df=5), aes(color = 'Chi(5)'), size = 1) +
    stat_function(fun = dchisq, args = c(df=8), aes(color = 'Chi(8)'), size = 1) + 
    labs(x = "x", y = "f(x)", color = "distribution")
```

```r
options(repr.plot.width=6, repr.plot.height=3)

ggplot(NULL, aes(c(0,15)))+
    stat_function(fun = pchisq, args = c(df=2), aes(color = 'Chi(2)'), size = 1) + 
    stat_function(fun = pchisq, args = c(df=3), aes(color = 'Chi(3)'), size = 1) +
    stat_function(fun = pchisq, args = c(df=5), aes(color = 'Chi(5)'), size = 1) +
    stat_function(fun = pchisq, args = c(df=8), aes(color = 'Chi(8)'), size = 1) + 
    labs(x = "x", y = "f(x)", color = "distribution")
```

## 3. Распределение Стьюдента

```r
options(repr.plot.width=6, repr.plot.height=3)

ggplot(NULL, aes(c(-5,5)))+
    stat_function(fun = dnorm, aes(color = 'N(0,1)'), size = 1) + 
    stat_function(fun = dt, args = c(df=1), aes(color = 't(1)'), size = 1,linetype = "dashed") +
    stat_function(fun = dt, args = c(df=2), aes(color = 't(2)'), size = 1,linetype = "dashed") +
    stat_function(fun = dt, args = c(df=5), aes(color = 't(5)'), size = 1,linetype = "dashed") + 
    stat_function(fun = dt, args = c(df=100), aes(color = 't(100)'), size = 1,linetype = "dashed") +
    labs(x = "x", y = "f(x)", color = "distribution")
```

```r
options(repr.plot.width=6, repr.plot.height=3)

ggplot(NULL, aes(c(-5,5)))+
    stat_function(fun = pnorm, aes(color = 'N(0,1)'), size = 1) + 
    stat_function(fun = pt, args = c(df=1), aes(color = 't(1)'), size = 1,linetype = "dashed") +
    stat_function(fun = pt, args = c(df=2), aes(color = 't(2)'), size = 1,linetype = "dashed") +
    stat_function(fun = pt, args = c(df=5), aes(color = 't(5)'), size = 1,linetype = "dashed") + 
    stat_function(fun = pt, args = c(df=100), aes(color = 't(100)'), size = 1,linetype = "dashed") +
    labs(x = "x", y = "F(x)",  color = "distribution")
```

## 4. Распределение Фишера

```r
options(repr.plot.width=6, repr.plot.height=3)

ggplot(NULL, aes(c(0,5)))+
    stat_function(fun = df, args = c(df1=3, df2=1), aes(color = 'F(3,1)'), size = 1) + 
    stat_function(fun = df, args = c(df1=5, df2=1), aes(color = 'F(5,1)'), size = 1) +
    stat_function(fun = df, args = c(df1=40, df2=1), aes(color = 'F(40,1)'), size = 1) +
    stat_function(fun = df, args = c(df1=20, df2=3), aes(color = 'F(20,3)'), size = 1) + 
    stat_function(fun = df, args = c(df1=20, df2=5), aes(color = 'F(20,2)'), size = 1) + 
    stat_function(fun = df, args = c(df1=20, df2=10), aes(color = 'F(20,10)'), size = 1) + 

    labs(x = "x", y = "f(x)", color = "distribution")
```

## 5. Проблема тяжёлых хвостов
