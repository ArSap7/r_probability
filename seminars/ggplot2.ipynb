```r
# Load the ggplot2 package
library(ggplot2)
library('repr')
options(repr.plot.width=4, repr.plot.height=3)

# Explore the mtcars data frame with str()
# str(mtcars)

# Execute the following command
ggplot(mtcars, aes(x = cyl, y = mpg)) +
  geom_point()
```

```r
# Add geom_point() with +
ggplot(diamonds, aes(x = carat, y = price)) + geom_point()


# Add geom_point() and geom_smooth() with +
ggplot(diamonds, aes(x = carat, y = price)) + geom_point() + geom_smooth()
```

```r
# 3 - Copy the above command and assign the correct value to col in aes()
ggplot(diamonds, aes(x = carat, y = price, color=clarity)) +
  geom_smooth()


# 4 - Keep the color settings from previous command. Plot only the points with argument alpha.
ggplot(diamonds, aes(x = carat, y = price, color=clarity)) +
  geom_point(alpha=0.4)
```

```r
# 1 - The dia_plot object has been created for you
dia_plot <- ggplot(diamonds, aes(x = carat, y = price))

# 2 - Expand dia_plot by adding geom_point() with alpha set to 0.2
dia_plot <- dia_plot + geom_point(alpha = 0.2)

# 3 - Copy the command from above and add aes() with the correct mapping to geom_smooth()
dia_plot + geom_smooth(aes(col = clarity), se = FALSE)
```

```r
mtcars$cyl <- as.factor(mtcars$cyl) # без as factor не будет рисовать нихера

ggplot(mtcars, aes(x = wt, y = mpg, col = cyl)) +
  geom_point( ) + 
  geom_smooth(method = 'lm', se = FALSE) + 
  geom_smooth(aes(group = 1), method = 'lm', se = FALSE, linetype = 2) 
# the group aesthetic will tell ggplot() to draw a single linear model through all the points.
```

```r
cyl.am <- ggplot(mtcars, aes(x = factor(cyl), fill = factor(am)))

# The base layer, cyl.am, is available for you
# Add geom (position = "stack" by default)
cyl.am + geom_bar( )

# Fill - show proportion
cyl.am + 
  geom_bar(position = 'fill')  

# Dodging - principles of similarity and proximity
cyl.am +
  geom_bar(position = "dodge") 

# Clean up the axes with scale_ functions
val = c("#E41A1C", "#377EB8")
lab = c("Manual", "Automatic")
cyl.am +
  geom_bar(position = "dodge") +
  scale_x_discrete("Cylinders") + 
  scale_y_continuous("Number") +
  scale_fill_manual("Transmission", 
                    values = val,
                    labels = lab) 
```
